```python
# 1. How many distinct (different) actors' last names are there?
query1 = 'SELECT COUNT(DISTINCT(last_name)) FROM sakila.actor'
q1 = pd.read_sql_query(query1, engine) #We use panda to read the query and then put in a SQL query
q1.head()
```

```python
# 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
query2 = 'SELECT COUNT(DISTINCT(last_name)) FROM sakila.actor'
q2 = pd.read_sql_query(query2, engine) #We use panda to read the query and then put in a SQL query
q2.head()
```

```python
# 3. Get all films with ARMAGEDDON in the title.
query3 = 'Select * \
FROM film \
WHERE title LIKE "%%ARMAGEDDON%%"'
q3 = pd.read_sql_query(query3, engine)
q3
```

```python
# 4. Get 10 the longest films.

query4 = ('SELECT * \
FROM film \
ORDER BY length DESC \
LIMIT 10;')
q4 = pd.read_sql_query(query4, engine)
q4
```

```python
# 5. How many films include Behind the Scenes content?

query5 = 'SELECT COUNT(title) AS "NO_behind_the_scenes" \
FROM film \
WHERE special_features LIKE "%%Behind the Scenes%%"'
q5 = pd.read_sql_query(query5, engine) 
q5
```

```python
# 6. Which kind of movies (rating) have a mean duration of more than two hours?
query6 = 'SELECT rating, ROUND(AVG(length),2) AS "mean_length" \ 
FROM film \
GROUP BY rating \
HAVING ROUND(AVG(length),2) > 120'
q6 = pd.read_sql_query(query6, engine) 
q6
```

```python
# 7. Rank films by length (filter out the rows that have nulls or 0s in length column). 
# In your output, only select the columns title, length, and the rank.

query7 = 'SELECT title, length, RANK() OVER(ORDER BY length) AS "ranking" \
from film \
WHERE length <> ' ' OR 0;'
q7 = pd.read_sql_query(query7, engine) 
q7
```
